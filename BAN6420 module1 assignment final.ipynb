##### Introduction:
   ###### In this assignment, we are going to create a weekly payment slip for the employees of highridge construction company. We will create a dynamic list of workers, use loop to generate slips for these workers and use conditional statement to properly configure the list.

```python
# we will import random package.

import random
```

```python

# first, we will list out what we is given, that is the constants and assign variables.

salary_range= (1000, 50000)
gender= ['male', 'female']
no_employees = 500
paylist=[]
```

```python
#we will create a function that contains our list of workers.

def employees(no_employees):
     
    for i in range(no_employees):
        
        Gender = random.choice(gender) #randomly assigning gender 
        
        try:                           #use error handling to point out common error in the salary range
            Salary = random.randint(*salary_range) #randomly assign salaries
        except TypeError:
            return "Adjust salary_range"
       
    #filling the level using conditional statement
        level  = 'NA'                          
        if Salary > 10000 and Salary < 20000:
            level = 'A1'
            
        elif (Salary > 7500 and Salary < 30000) and Gender == "female":
            level = 'A5-F'
             
        paylist.append({'id': i+1, 'Salary':f"${Salary:,}", 'Gender': Gender, 'Level': level}) #create the list

        
    return paylist
```

```python
#generating payslip

final_paylist= employees(no_employees)
```
